```python
import scrapy
import pandas as pd
```

```python
urls = []
for i in range(1,666):
    urls.append('https://indeks.kompas.com/?site=otomotif&page='+str(i))
```

```python
class Berita(scrapy.Spider):
    name = 'berita_spider'

    custom_settings = {
        'FEED_FORMAT' : 'csv',
        'FEED_URI' : 'berita_kompas_otomotif_url.csv' 
    }
    def start_requests(self):
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)
    def parse(self, response):
        url = response.xpath('//a[contains(@class,"article__link")]/@href').extract()
        title = response.xpath('//a[contains(@class,"article__link")]/text()').extract()
        for item in zip(title,url):          
            scraped_info = {
                'title': item[0],
                'url': item[1]
            }
            yield scraped_info
```

```python
from scrapy.crawler import CrawlerProcess
process = CrawlerProcess()
process.crawl(Berita)
process.start()
```

```python
class BeritaDetail(scrapy.Spider):
    name = 'berita_spider'
    
#     custom settings is use to define some certain scrapy settings in the process of running the spider.
#     the settings we use is FEED_FORMAT to define the crawled data format, 
#     and FEED_URI to define the name of the crawled data file.
    custom_settings = {
        'FEED_FORMAT' : 'csv',
        'FEED_URI' : 'berita_kompas_tekno.csv' 
    }
    
#     define the start_requests function that led to our web url that we're going to crawl.
    def start_requests(self):
        urls_detail =['https://tekno.kompas.com/read/2020/10/01/09420067/google-maps-versi-gelap-mulai-disebar-di-android']
        #urls = ['https://www.gramedia.com/search?based_on=best-seller&category=buku']
        #urls = ['http://gmdki2.xyz/']
        for url in urls_detail:
            yield scrapy.Request(url=url, callback=self.parse)

#     define the parse function that navigates to the specific contents of the web that we're going to crawl
#     we use css locator for this example
    def parse(self, response):
        content = response.xpath('//div[contains(@class,"read__content")]/p/text()').getall()
        label = 'tekno'
        #uthor = response.xpath('//a[contains(@class,"authorName")]/span/text()').extract()
        #ating = response.xpath('//span[contains(@class,"minirating")]/text()')
        #ating = [p.replace('avg','').replace(',','.').replace('ratings','').replace(' ','').replace('—',' ').replace('rating','') for p in rating.extract()]
        for item in zip(content,label):          
            scraped_info = {
                'content': item[0],
                'label': item[1]
            }
            yield scraped_info
```

```python
from scrapy.crawler import CrawlerProcess
process = CrawlerProcess()
process.crawl(BeritaDetail)
process.start()
```
