```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
factory_stemming = StemmerFactory()
stemmer = factory_stemming.create_stemmer()
postagger = PosTag()
import re
import string  
def remove_punct(text):
    text_nonpunct = ''
    text_nonpunct = re.sub('['+string.punctuation+']','',text)
    return text_nonpunct

def lower_token(tokens):
    return [w.lower() for w in tokens]

def cleanParagraph(paragraph):
    paragraph_stopword_removed = stopword.remove(paragraph)
    text_steammed = stemmer.stem(paragraph_stopword_removed)
    return text_steammed

def generateCleanText(lists):
    clean_text = []
    length_data = len(lists['content'])
    for i in range(0,length_data):
        lists['content'][i] = BeautifulSoup(lists['content'][i], "lxml").text
        lists['content'][i] = lists['content'][i].replace('Liputan6.com, ','').replace('JAKARTA,KOMPAS.com','').replace('JAKARTA, KOMPAS.com','').replace('Liputan.com, Jakarta','')
        lists['content'][i] = lists['content'][i].replace('kompas com','')
        lists['content'][i] = re.sub(r"\d+","",lists['content'][i])
        lists['content'][i] = re.sub('[^a-zA-Z]', ' ',lists['content'][i])
        lists['content'][i] = lists['content'][i].replace('\n','')
        lists['content'][i] = lists['content'][i].lower()
        clean_text.append(cleanParagraph(lists['content'][i]))
    return clean_text

#def tokenizing(listData):
 #   tokens = []
 #   length_data = len(listData['text_clean'])
 #   for i in range(0,length_data):
#        listData['text_clean'][i] = remove_punct(listData['text_clean'][i])
#        tokens.append(word_tokenize(listData['text_clean'][i]))
#    return tokens
def tokenizing(listData):
    tokens = []
    length_data = len(listData['text_clean'])
    for i in range(0,length_data):
        listData['text_clean'][i] = remove_punct(listData['text_clean'][i])
        tokens.append(postagger.get_pos_tag(listData['text_clean'][i]))
    return tokens
def encodingLabel(listData):
    bola = []
    news = []
    bisnis = []
    tekno = []
    otomotif = []
    for l in listData.label:
        if l == 'news':
            bola.append(0)
            news.append(1)
            bisnis.append(0)
            tekno.append(0)
            otomotif.append(0)
        elif l == 'bola':
            bola.append(1)
            news.append(0)
            bisnis.append(0)
            tekno.append(0)
            otomotif.append(0)
        elif l == 'bisnis':
            bola.append(0)
            news.append(0)
            bisnis.append(1)
            tekno.append(0)
            otomotif.append(0)
        elif l == 'tekno':
            bola.append(0)
            news.append(0)
            bisnis.append(0)
            tekno.append(1)
            otomotif.append(0)
        elif l == 'otomotif':
            bola.append(0)
            news.append(0)
            bisnis.append(0)
            tekno.append(0)
            otomotif.append(1)

    listData['bola'] = bola
    listData['news'] = news
    listData['bisnis'] = bisnis
    listData['tekno'] = tekno
    listData['otomotif'] = otomotif
    return listData
```

```python
from gensim.models.fasttext import FastText
fasttext = FastText.load_fasttext_format('E:\FastText BIN\cc.id.300.bin')
```

```python
MAX_SEQUENCE_LENGTH = 150
tokenizer = Tokenizer(num_words=len(TRAINING_VOCAB),filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~',lower=True,char_level=False)
tokenizer.fit_on_texts(d_train["text_clean"].values)
training_sequences = tokenizer.texts_to_sequences(d_train["text_clean"].values)

train_word_index = tokenizer.word_index
print('Found %s unique tokens.'%len(train_word_index))
train_cnn_data = pad_sequences(training_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding="post")
```

```python
test_sequences = tokenizer.texts_to_sequences(d_test["text_clean"].values)
test_cnn_data = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH,padding="post")
print('Shape of data tensor:', test_cnn_data.shape)
```

```python
average = sum(training_sentence_lengths)/len(training_sentence_lengths)
average
```

```python
EMBEDDING_DIM = 300

train_embedding_weights = np.zeros((len(train_word_index)+1, EMBEDDING_DIM))
for word,index in train_word_index.items():
    train_embedding_weights[index,:] = fasttext[word] if word in fasttext else np.random.rand(EMBEDDING_DIM)
print(train_embedding_weights.shape)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
def build_lstm():
    lstm_out = 150
    dropout_rate = 0.5
    max_features = train_embedding_weights.shape[0]
    model = Sequential()
    model.add(Embedding(max_features,EMBEDDING_DIM,weights=[train_embedding_weights],input_length=train_cnn_data.shape[1], dropout=dropout_rate))
    model.add(SpatialDropout1D(dropout_rate))
    model.add(LSTM(lstm_out,dropout=dropout_rate,recurrent_dropout=dropout_rate))
    model.add(Dense(5,activation='softmax'))
    model.compile(loss='categorical_crossentropy',optimizer="adam",metrics=['accuracy'])
    return model
classifier = KerasClassifier(build_fn=build_lstm,batch_size=128,epochs=21)
```

```python
label_names = ['news', 'bola','bisnis','otomotif','tekno']
y_train = d_train[label_names].values
x_train = train_cnn_data
y_tr = y_train
y_te = d_test[label_names].values
x_test = test_cnn_data
y_test = y_te
```

```python
#lstm - rnn
import time
start_time = time.time()
num_epochs = 21
batch_size = 128
hist = model.fit(x_train,
                 y_tr, 
                 epochs = num_epochs, 
                 validation_data=(x_test, y_test),
                 shuffle=True,
                 batch_size=batch_size,
                 )
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
plt.title('Loss')
plt.plot(hist.history['loss'], label='train')
plt.plot(hist.history['val_loss'], label='test')
plt.legend()
plt.show();
```

```python
plt.title('Accuracy')
plt.plot(hist.history['accuracy'], label='train')
plt.plot(hist.history['val_accuracy'], label='test')
plt.legend()
plt.show();
```
